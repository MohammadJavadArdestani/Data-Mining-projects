```python
import sklearn.datasets
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

# Evaluating NN's Elements 

```python
X, y = sklearn.datasets.make_circles(n_samples=2000, shuffle=True, noise=0.2, random_state=10, factor=0.8)
```

```python
plt.scatter(X[:,0], X[:, 1], c = y)
```

## Without Activation Function 



```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation = None))
model.add(tf.keras.layers.Dense(32, activation = None))
model.add(tf.keras.layers.Dense(10, activation = None))
model.add(tf.keras.layers.Dense(1, activation = None))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# This builds the model for the first time:
model.fit(X, y, epochs=10, verbose = 0)
loss, accuracy = model.evaluate(X, y)
print(f' Model loss on the test set: {loss}')
print(f' Model accuracy on the test set: {100*accuracy}')
```

As we see, the model couldn't have a good accuracy without an activation function. Activation functions help us to classify datasets that are not linear. 

## Linear Activation Function 

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(32, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(10, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(2, activation = tf.keras.activations.linear))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# This builds the model for the first time:
model.fit(X, y, epochs=10, verbose = 0)
loss, accuracy = model.evaluate(X, y)
print(f' Model loss on the test set: {loss}')
print(f' Model accuracy on the test set: {100*accuracy}')
```

### Linear activation couldn't help us because we don't have a linear dataset for this problem. 

## Regression Loss

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(32, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(10, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(2, activation = None))
model.compile(loss= "MeanSquaredError", optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose = 0)
loss, accuracy = model.evaluate(X, y)
print(f' Model loss on the test set: {loss}')
print(f' Model accuracy on the test set: {100*accuracy}')
```

###  As a rule of thumb, mean squared error (MSE) is more appropriate for regression problems, that is, problems where the output is a numerical value but sometimes using the MSE is not a bad idea. In this problem, the loss function is calculated by the probability of membership data into each class. Although it wasn't helpful for our situation,  the accuracy doesn't fall too much. 

## One hidden Layer Network

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10, activation = tf.keras.activations.linear))
model.add(tf.keras.layers.Dense(1, activation = tf.keras.activations.linear))
model.compile(loss= "binary_crossentropy", optimizer='adam', metrics=['accuracy'])
# This builds the model for the first time:
model.fit(X, y, epochs=10, verbose = 0)
loss, accuracy = model.evaluate(X, y)
print(f' Model loss on the test set: {loss}')
print(f' Model accuracy on the test set: {100*accuracy}')
```

### one hidden layer couldn't learn alot about the dataset

## Learning Rate

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation = 'relu'))
model.add(tf.keras.layers.Dense(32, activation = 'relu'))
model.add(tf.keras.layers.Dense(10, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
model.compile(loss= "binary_crossentropy", optimizer = tf.keras.optimizers.Adam(lr = 0.8), metrics=['accuracy'])
model.fit(X, y, epochs = 10)
```

If we have large learning rate, the loss value would be unstable and couldn't keep decreasing. 

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation = 'relu'))
model.add(tf.keras.layers.Dense(32, activation = 'relu'))
model.add(tf.keras.layers.Dense(10, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
model.compile(loss= "binary_crossentropy", optimizer = tf.keras.optimizers.Adam(lr = 0.00001), metrics=['accuracy'])
model.fit(X, y, epochs = 10)
```

### Very small learning rata doesn't make learning very slow, and we should increase epochs to avoid underfitting. 

## Best results

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, activation = 'relu'))
model.add(tf.keras.layers.Dense(16, activation = 'relu'))
model.add(tf.keras.layers.Dense(8, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
model.compile(loss= "binary_crossentropy", optimizer = tf.keras.optimizers.Adam(lr = 0.005), metrics=['accuracy'])
model.fit(X, y, epochs = 20, verbose = 0)
loss, accuracy = model.evaluate(X, y)
print(f' Model loss on the test set: {loss}')
print(f' Model accuracy on the test set: {100*accuracy}')
```

### Load and normalize data

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images = train_images / 255.0
test_images = test_images / 255.0
```

### Explore Data

```python
print(f"number of class: {len(set(test_labels))}")
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
utils.plot_multi_images(train_images, train_labels, class_names)
```

### Define Model

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
model.add(tf.keras.layers.Dense(150, activation='relu'))
model.add(tf.keras.layers.Dense(75, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
model.compile(loss= "sparse_categorical_crossentropy", optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
              metrics=['accuracy'])
```

### Train and Validate Model

```python
model.fit(train_images, train_labels, validation_split=0.25, epochs=7)
```

### Test Model

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test accuracy:', test_acc)
```

### Confusion Matrix

```python
predictions = model.predict(test_images)
```

```python
predicted_classes = utils.label_with_highest_prob(predictions)
utils.plot_confusion_matrix(y_pred=predicted_classes, y_true=test_labels, classes=np.array(class_names))
```
